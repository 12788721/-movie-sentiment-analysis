```python
# ===============================
# STEP 4 : Text Preprocessing
# ===============================

import string
from bs4 import BeautifulSoup
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):

    # Convert to lowercase
    text = text.lower()

    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove stopwords
    words = text.split()

    words = [word for word in words if word not in stop_words]

    return " ".join(words)
```

```python
print("Cleaning reviews...")

df['clean_review'] = df['review'].apply(clean_text)

print("Cleaning Completed!")
```

```python
# ===============================
# Features and Target
# ===============================

X = df['clean_review']

y = df['sentiment']

print("Features Shape:", X.shape)
print("Labels Shape:", y.shape)
```

```python
# ===============================
# Train-Test Split
# ===============================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Dataset Split Successfully!")
```

```python
print("Training Reviews :", X_train.shape)
print("Testing Reviews  :", X_test.shape)

print()

print("Training Labels :", y_train.shape)
print("Testing Labels  :", y_test.shape)
```

```python
print("Training Labels Distribution:")
print(y_train.value_counts())

print("\nTesting Labels Distribution:")
print(y_test.value_counts())
```

```python
# ===============================
# STEP 7: TF-IDF Feature Extraction
# ===============================

tfidf = TfidfVectorizer(
    max_features=50000,
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.95,
    sublinear_tf=True
)

print("TF-IDF Vectorizer Created Successfully!")
```

```python
# Learn vocabulary and transform training data

X_train_tfidf = tfidf.fit_transform(X_train)

print("Training Data Vectorized Successfully!")
```

```python
# Transform testing data using the same vocabulary

X_test_tfidf = tfidf.transform(X_test)

print("Testing Data Vectorized Successfully!")
```

```python
print("Training Matrix Shape:", X_train_tfidf.shape)
print("Testing Matrix Shape :", X_test_tfidf.shape)
```

```python
print("Vocabulary Size:", len(tfidf.vocabulary_))
```

```python
joblib.dump(tfidf, "tfidf_vectorizer.pkl")

print("TF-IDF Vectorizer Saved Successfully!")
```

```python
# ===============================
# STEP 8: Create Linear SVM Model
# ===============================

svm_model = LinearSVC(
    C=1.0,
    random_state=42,
    max_iter=3000
)

print("Linear SVM Model Created Successfully!")
```

```python
# ===============================
# Train the Model
# ===============================

print("Training Linear SVM...")

svm_model.fit(X_train_tfidf, y_train)

print("Training Completed Successfully!")
```

```python
# ===============================
# Make Predictions
# ===============================

y_pred = svm_model.predict(X_test_tfidf)

print("Predictions Completed!")
```

```python
comparison = pd.DataFrame({

    "Actual": y_test.values,

    "Predicted": y_pred

})

comparison.head(10)
```

```python
# ===============================
# Save the Model
# ===============================

joblib.dump(svm_model, "linear_svm_model.pkl")

print("Model Saved Successfully!")
```

```python
# ===============================
# STEP 9: Model Evaluation
# ===============================

# Calculate Accuracy

accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
# Classification Report

print("Classification Report\n")

print(classification_report(y_test, y_pred))
```

```python
# Confusion Matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Negative', 'Positive'],
    yticklabels=['Negative', 'Positive']
)

plt.title("Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.5, 1, 2, 5]
}
```

```python
grid_search = GridSearchCV(

    estimator=LinearSVC(max_iter=3000),

    param_grid=param_grid,

    cv=5,

    scoring='accuracy',

    n_jobs=-1,

    verbose=2

)
```

```python
print("Starting Hyperparameter Tuning...")

grid_search.fit(
    X_train_tfidf,
    y_train
)

print("Hyperparameter Tuning Completed!")
```

```python
print("Best Parameter:")

print(grid_search.best_params_)
```

```python
print("Best Cross Validation Accuracy:")

print(grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
```

```python
y_pred_best = best_model.predict(X_test_tfidf)
```

```python
accuracy_best = accuracy_score(
    y_test,
    y_pred_best
)

print(f"Final Accuracy: {accuracy_best*100:.2f}%")
```

```python
print(classification_report(
    y_test,
    y_pred_best
))
```
